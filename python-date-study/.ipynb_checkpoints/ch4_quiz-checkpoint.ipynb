# books.json 파일을 불러오세요.

가격이 20000원 이상인 책만 출력하세요.<br>

가격이 비싼 순으로 정렬하여 출력하세요.<br>

```python
books = books.set_index('제목')
books
```

```python
# 가격이 20000원 이상인 책만 출력하세요.
books[books['가격'] >= 20000]
```

```python
# 대출가능권수가 2권 이하인 책을 출력하세요.
books[books['대출가능권수'] <= 2]
```

```python
# 가격이 비싼 순으로 정렬하여 출력하세요.
books.sort_values('가격', ascending=False)
```

```python
# 대출가능권수가 많은 순으로 정렬하여 출력하세요.
books.sort_values('대출가능권수', ascending=False)
```

```python
# 전체 평균 가격을 구하세요.
print('전체 평균 가격 :', int(books['가격'].mean()))
```

# 공공 데이터 포털에서 인천 정류장별 이용승객 현황을 검색하세요.
csv 파일을 다운받고, 파이썬 프로그램으로 불러오세요.<br><br>

총승차건수가 10만 이상인 정류장을 출력하세요.<br>
총하차건수가 0건인 정류장을 출력하세요.<br>
승차건수(현금)이 100건 이상인 정류장을 출력하세요.<br>
승차건수(현금)이 승차건수(카드)보다 많은 정류장을 출력하세요.

```python
bus = pd.read_csv('quiz/인천광역시_정류장별 이용승객 현황_20250731.csv', encoding='euc-kr')
bus
```

```python
# 총승차건수가 10만 이상인 정류장을 출력하세요.
bus[bus['총승차건수'] >= 100000]
```

```python
# 총하차건수가 0건인 정류장을 출력하세요.
bus[bus['총하차건수'] == 0]
```

```python
# 승차건수(현금)이 100건 이상인 정류장을 출력하세요.
bus[bus['승차건수(현금)'] >= 100]
```

```python
# 승차건수(현금)이 승차건수(카드)보다 많은 정류장을 출력하세요.
bus[bus['승차건수(현금)'] > bus['승차건수(카드)']]
```

# 공공 데이터 포털에서 인천광역시 중구 공영주차장을 검색하세요.
csv 파일을 다운받고, 파이썬 프로그램으로 불러오세요.<br><br>

주차구획수(주차 가능한 차량 대수)가 100대 이상인 주차장을 모두 출력하세요.<br>
평일운영종료시각이 21:00인, 주차장을 출력하세요.<br>
요금정보가 무료인 주차장을 모두 출력하세요.<br>
모든 주차장의 주차구획수 합계를 구하고, 평균도 구하세요.

```python
parking = pd.read_csv('quiz/인천광역시 중구_공영주차장 현황_20250601.csv', encoding='utf-8')
parking
```

```python
# 주차구획수(주차 가능한 차량 대수)가 100대 이상인 주차장을 모두 출력하세요.
parking[parking['주차구획수'] >= 100]
```

```python
# 평일운영종료시각이 21:00인, 주차장을 출력하세요.
parking[parking['평일운영종료시각'] == '21:00']
```

```python
# 요금정보가 무료인 주차장을 모두 출력하세요.
parking[parking['요금정보'] == '무료']
```

```python
# 모든 주차장의 주차구획수 합계를 구하고, 평균도 구하세요.
su = int(parking['주차구획수'].sum())
avg = int(parking['주차구획수'].mean())
print(f'모든 주차장의 주차구획수 합 : {su}, 평균 : {avg}')
```
