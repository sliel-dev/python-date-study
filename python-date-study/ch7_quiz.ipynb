# 막대 그래프를 그리세요.
과일 : ['사과', '바나나', '포도', '오렌지']
판매량 : [50, 80, 40, 70]

```python
a = ['사과', '바나나', '포도', '오렌지']
b = [50, 80, 40, 70]
plt.bar(a, b)
plt.xlabel('과일')
plt.ylabel('판매량')
```

# 산점도 그래프를 그리세요.

```python
x = [1, 2, 3, 4, 5] 
y = [40, 55, 65, 80, 85]
plt.scatter(x, y)
plt.xlabel('공부시간')
plt.ylabel('시험점수')
```

# 히스토그램을 그리세요.
- 값: 하루 유튜브 시청 시간(0~10시간 사이, 100명)

```python
day = np.random.randint(0, 11, 100)
plt.hist(day, bins=10)
plt.title('하루 유튜브 시청 시간 분포')
plt.xlabel('시청 시간(HH)')
plt.ylabel('사람 수')
```

# 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
데이터 : 과일 판매량
과일 : ['사과', '바나나', '포도', '오렌지', '복숭아']
판매 개수 : [50, 80, 40, 70, 60]

```python
fruit = ['사과', '바나나', '포도', '오렌지', '복숭아']
ea = [50, 80, 40, 70, 60]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].plot(ea)
axs[1].bar(fruit, ea)
```

# 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
데이터 : 요일별 유튜브 시청 시간
요일 : ['월', '화', '수', '목', '금']
시청 시간 : [2, 3, 4, 2, 5]

```python
data = ['월', '화', '수', '목', '금']
tim = [2, 3, 4, 2, 5]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].plot(tim)
axs[1].bar(data, tim)
```

# 하나의 도화지에 2개의 그래프(막대 그래프, 산점도) 그리세요.
데이터 : 요일별 독서 시간
요일 : ['월', '화', '수', '목', '금’]
독서 시간 : [1, 0.5, 2, 1.5, 2.5]

```python
data = ['월', '화', '수', '목', '금']
readti = [1, 0.5, 2, 1.5, 2.5]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].bar(data, readti)
axs[1].scatter(data, readti)
```

# 하나의 도화지에 2개의 그래프(선 그래프, 산점도)를 그리세요.
데이터 : 도시별 평균 기온(°C)
도시 : ['서울', '부산', '대구', '광주', '제주’]
평균 기온 : [15.5, 17.8, 16.2, 15.0, 18.3]

```python
city_temp = ['서울', '부산', '대구', '광주', '제주']
avg_temp = [15.5, 17.8, 16.2, 15.0, 18.3]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].plot(city_temp, avg_temp)
axs[1].scatter(city_temp, avg_temp)
```
