```python
fig = plt.figure(figsize=(10, 3))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
students = [1, 2, 3, 4, 5]
scores = [80, 90, 70, 85, 95]
ax1.plot(scores)
ax2.bar(students, scores)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
students = [1, 2, 3, 4, 5]
scores = [80, 90, 70, 85, 95]
axs[0].plot(scores)
axs[1].bar(students, scores)
```
