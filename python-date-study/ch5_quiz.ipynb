# 서울시설공단_서울도시고속도로 노선별 시간대별 교통량.csv 파일을 불러오세요.
노선별로 그룹핑하여 교통량 합계를 구하세요.

```python
seoul_s = pd.read_csv('0917/서울시설공단_서울도시고속도로 노선별 시간대별 교통량.csv', encoding='euc-kr')
seoul_s = seoul_s.dropna()
seoul_s = seoul_s.astype({'시간대':'int64', '교통량':'int64'})
```

```python
# 노선별로 그룹핑하여 교통량 합계를 구하세요
seoul_s_not_gr = seoul_s.groupby('노선').sum()
seoul_s_not_gr = seoul_s_gr.astype({'교통량' : 'int64'})
seoul_s_not_gr['교통량']
```

```python
# 노선별로 그룹핑하여 교통량 평균을 구하세요
seoul_s.groupby('노선')['교통량'].mean()
```

```python
# 방향별 총 교통량
seoul_s_si_gr = seoul_s.groupby('방향').sum()
seoul_s_si_gr = seoul_s_si_gr.astype({'교통량':'int64'})
seoul_s_si_gr['교통량']
```

```python
# 시간대별 평균 교통량
seoul_s_ti_gr = seoul_s.groupby('시간대')['교통량'].mean()
seoul_s_ti_gr
```

```python
# 교통량이 가장 많은 시간대
seoul_s_ti_gr = seoul_s.groupby('시간대').sum()
seoul_s_ti_gr['교통량']
```

# 서울시 버스노선별 정류장별 승하차 인원 정보.csv 파일을 불러오세요.

```python
seoul_bus = pd.read_csv('0917/서울시 버스노선별 정류장별 승하차 인원 정보.csv', encoding='euc-kr')
seoul_bus
```

```python
# 노선번호별 총 승차승객 수
seoul_bus_no_gr = seoul_bus.groupby('노선번호').sum()
seoul_bus_no_gr['승차총승객수']
```

```python
# 노선번호별 총 하차승객 수
seoul_bus_no_gr['하차총승객수']
```

```python
# 741 노선이 지나가는 정류장별 승차승객 수
seoul_bus_no_gr = seoul_bus.groupby('노선명').sum()
seoul_bus_no_gr[seoul_bus_no_gr['노선번호'].str.contains('741')]['승차총승객수']
```

# 코로나 확진자 현황 데이터

연도별 총 확진자수<br>
동별 평균 확진자수<br>
2021년도 월별 확진자수의 합계<br>
확진자수가 없는 행의 개수<br>
동별 확진자수를 구하고, 큰값 순으로 정렬<br>
2022년 확진자수가 가장 많은 동과 확진자수<br>

```python
data = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding='euc-kr')
data
```

```python
# 연도별 총 확진자수
data.groupby('기준연도').sum()['확진자수']
```

```python
# 동별 평균 확진자수
data.groupby('법정동')['확진자수'].mean()
```

```python
# 2021년도 월별 확진자수의 합계
c_2021 = data.groupby(['기준연도', '기준월']).sum()['확진자수'].loc[2021]
c_2021
```

```python
# 확진자수가 없는 행의 개수
data['확진자수'].isnull().value_counts()
```

```python
# 동별 확진자수를 구하고, 큰값 순으로 정렬
dong_zombie = data.groupby('법정동')['확진자수'].sum()
dong_zombie.sort_values(ascending=False)
```

```python
# 2022년 확진자수가 가장 많은 동과 확진자수
c_2022 = data.groupby(['기준연도', '법정동']).sum()['확진자수'].loc[2022]
c_2022.sort_values(ascending=False).head(1)
```
